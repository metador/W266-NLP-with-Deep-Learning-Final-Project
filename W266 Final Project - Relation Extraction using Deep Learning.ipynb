#### Create necessary functions i.e. word emeddings, CNN, PCNN, softmax, Logit and Output functions

```python
tf.reset_default_graph()



def network_embedding(train_loader,data_word,pos1,pos2, include_unk_blank = True,max_length = 120, position_dim = 5):
    with tf.variable_scope("word_embeddings",reuse=tf.AUTO_REUSE):
        word_embedding = tf.get_variable('word_embedding',initializer=train_loader.word_vec_mat,dtype=tf.float32)
        if include_unk_blank:
            word_embedding = tf.concat([word_embedding,tf.get_variable('unk_word_embedding',shape = [1,train_loader.word_vec_dim],\
                                        initializer=tf.contrib.layers.xavier_initializer(),dtype=tf.float32),\
                                        tf.constant(np.zeros((1, train_loader.word_vec_dim), dtype=np.float32))], 0)
    x = tf.nn.embedding_lookup(word_embedding,data_word)
    #print(x.shape)

    with tf.variable_scope('Positional_embeddings', reuse=tf.AUTO_REUSE):
        ## Creating a positional embedding 
        position_embedding_1 = tf.get_variable('positional_embedding_pos1',shape = [max_length*2,position_dim],\
                                            initializer=tf.contrib.layers.xavier_initializer(),dtype=tf.float32)
        position_embedding_2 = tf.get_variable('positional_embedding_pos2',shape = [max_length*2,position_dim],\
                                            initializer=tf.contrib.layers.xavier_initializer(),dtype=tf.float32)
        positions_1 = tf.nn.embedding_lookup(position_embedding_1,pos1)
        positions_2 = tf.nn.embedding_lookup(position_embedding_2,pos2)
        pos = tf.concat([positions_1,positions_2],-1)
        #print(pos.shape)
    x = tf.concat([x,pos],-1)
    #print(x.shape)
    return x

def conv_layer(h, hidden_size=230, kernel_size=3, stride_size=1,i=1):
    
    with tf.variable_scope("conv_layer", reuse=tf.AUTO_REUSE):
        # looping though various convoulted layers

        #for i,size in enumerate(hidden_dims):
        conv_layer = tf.layers.conv1d(inputs=h, \
                     filters=hidden_size,\
                     kernel_size=kernel_size,\
                     strides=stride_size, \
                     padding='same', \
                     kernel_initializer=tf.contrib.layers.xavier_initializer())
        #print(conv_layer.shape)
        return conv_layer
        
def cnn(h, hidden_dims=[230], kernel_size=[3], stride_size=[1], activation=tf.nn.relu, keep_prob=1.0,  is_training=False):
    
    with tf.variable_scope("conv_layers", reuse=tf.AUTO_REUSE):
        # looping though various convoulted layers
        if  len(hidden_dims) != len(kernel_size) and len(kernel_size) !=len(stride_size):
            raise Exception('Length of the list of hidden dimension, kernel size and stride size should be same')
        
        for i,size in enumerate(hidden_dims): ## function to have multiple layers of conv
            h = conv_layer(h, i=i, hidden_size= hidden_dims[i], kernel_size=kernel_size[i],\
                           stride_size=stride_size[i])
        print("Conv layer", h.shape)
        #pool = tf.reduce_max(h,axis=-2)
        pool = tf.layers.max_pooling1d(h,pool_size=int(h.shape[-2]), strides=1, padding='VALID')
        print("Pool shape", pool.shape)
        pool = tf.reshape(tf.squeeze(pool),tf.convert_to_tensor([-1,int(pool.get_shape()[-1])]))
        print("Pool shape after Squeeze", pool.shape)
        
        ## since tf.nn.relu or any activation funtion is activation
        activation_layer = activation ## for better readability
        x = activation_layer(pool)
        x = tf.layers.dropout(x, rate = 1-keep_prob , training = is_training)
        
    return x

def pcnn(h, mask, hidden_dims=[230], kernel_size=[3], stride_size=[1], activation=tf.nn.relu, keep_prob=1.0,  is_training=False):
    
    with tf.variable_scope("conv_layers", reuse=tf.AUTO_REUSE):
        # looping though various convoulted layers
        if  len(hidden_dims) != len(kernel_size) and len(kernel_size) !=len(stride_size):
            raise Exception('Length of the list of hidden dimension, kernel size and stride size should be same')
        
        for i,size in enumerate(hidden_dims):
            h = conv_layer(h, i=i, hidden_size= hidden_dims[i], kernel_size=kernel_size[i],\
                           stride_size=stride_size[i])
        print("conv layer", h.shape)
        #pool = tf.reduce_max(h,axis=-2)
        mask_embedding = tf.constant([[0, 0, 0], [1, 0, 0], [0, 1, 0], [0, 0, 1]], dtype=np.float32)
        mask = tf.nn.embedding_lookup(mask_embedding, mask)
        hidden_size = h.shape[-1]
        
        h = tf.reduce_max(tf.expand_dims(mask * 100, 2) + tf.expand_dims(h, 3), axis=1) - 100
        print('Pool after reduction' , h.shape)
        piecewise_pool = tf.reshape(h, [-1, hidden_size * 3])
        print('piecewise pool', piecewise_pool.shape)
        
        ## since tf.nn.relu or any activation funtion is activation
        activation_layer = activation ## for better readability
        x = activation_layer(piecewise_pool)
        x = tf.layers.dropout(x, rate = 1-keep_prob , training = is_training)
        
    return x

def get_logit(x, label,batch_size, rel_tot, weights_table=None, var_scope=None):
    with tf.variable_scope('W_out',reuse=tf.AUTO_REUSE):
        W_out = tf.get_variable('W_out', shape=[x.shape[1], rel_tot], dtype=tf.float32, initializer=tf.contrib.layers.xavier_initializer())
        b_out = tf.get_variable('bias', shape=[rel_tot], dtype=tf.float32, initializer=tf.contrib.layers.xavier_initializer())
        logit = tf.matmul( x,W_out) + b_out
    return logit
        

def get_output(x):
        return tf.argmax(x, axis=-1)
    
def softmax_cross_entropy(x, label, rel_tot, weights_table=None, var_scope=None):
    with tf.variable_scope(var_scope or "loss", reuse=tf.AUTO_REUSE):
        if weights_table is None:
            weights = 1.0
        else:
            weights = tf.nn.embedding_lookup(weights_table, label)
        label_onehot = tf.one_hot(indices=label, depth=rel_tot, dtype=tf.int32)
        loss = tf.losses.softmax_cross_entropy(onehot_labels=label_onehot, logits=x, weights=weights)
        tf.summary.scalar('loss', loss)
        #cross_entropy_ = tf.nn.sparse_softmax_cross_entropy_with_logits(logits = logits_, labels = labels_)
        #loss_ = tf.reduce_mean(cross_entropy_)

        return loss
    
def network_input(MODEL_CLASS,train_loader):
    pass
```

#### Creating tensorflow variables and creating the CNN Network

```python
tf.reset_default_graph()

word = tf.placeholder(dtype=tf.int32, shape=[None, max_length], name='word')
pos1 = tf.placeholder(dtype=tf.int32, shape=[None, max_length], name='pos1')
pos2 = tf.placeholder(dtype=tf.int32, shape=[None, max_length], name='pos2')
label = tf.placeholder(dtype=tf.int32, shape=[None], name='label')

print("#########################")
print("   Create Network graph")
print("#########################")

print("Printing shapes along the network for tracking the dimensions")
x =  network_embedding(train_loader, word,pos1,pos2, include_unk_blank = True,max_length = 120, position_dim = 5)
print("Embedding shape is given by", x.shape)
conv = cnn(x, hidden_dims=[230], kernel_size=[3], stride_size=[1], activation=tf.nn.relu, keep_prob=0.2,  is_training=is_training)
print('Convolutional Layer shape is given by ', conv.shape)
logit = get_logit(conv,label,batch_size = train_loader.batch_size,rel_tot = train_loader.rel_tot)
print('Logit layer shape is given by ',logit.shape)
loss = softmax_cross_entropy(logit,label, rel_tot = train_loader.rel_tot)
output = get_output(logit)
print('Output layer shape is given by ',output.shape)


# Optimizer

global_step = tf.Variable(0, name='global_step', trainable=False)
tf.summary.scalar('learning_rate', learning_rate)


with tf.variable_scope("Training", reuse=tf.AUTO_REUSE):
    # specify optimizer
    opt_ = tf.train.AdamOptimizer(learning_rate=learning_rate, )
    # compute ordinary gradients
    gradients_, variables = zip(*opt_.compute_gradients(loss))
    # clip the gradients, update them
    gradients_, _ = tf.clip_by_global_norm(gradients_, max_grad_norm_)
    # apply new gradients on optimizer
    train_op_ = opt_.apply_gradients(zip(gradients_, variables),global_step=global_step)
```

```python
def one_step(sess,batch_data,results_array=[]):
    batch_label = []
    feed_dict = {
        word: batch_data['word'],
        pos1: batch_data['pos1'],
        pos2: batch_data['pos2'],
        label: batch_data['rel']
    }
    is_training = True
    result = sess.run(results_array, feed_dict)
    batch_label.append(batch_data['rel'])
    batch_label = np.concatenate(batch_label)
    return result + [batch_label]

def one_step_test(sess,batch_data,results_array=[]):
    batch_label = []
    feed_dict = {
        word: batch_data['word'],
        pos1: batch_data['pos1'],
        pos2: batch_data['pos2'],
        label: batch_data['rel']
    }
    is_training = False
    result = sess.run(results_array, feed_dict)
    batch_label.append(batch_data['rel'])
    batch_label = np.concatenate(batch_label)
    return result + [batch_label]
```

### Running the graph. Skip to next cell to load the trained  model and avoid training

```python

sess = tf.Session()
sess.run(tf.global_variables_initializer())

best_test_loss = 10000000000000 # large vaule
count = 0
break_early = False

tot_rel = test_loader.rel_tot
val_batch = test_loader.next_batch(5000)
early_stopping = False
for epoch in range(max_epoch):
    if break_early:
        break
    print('###### Epoch ' + str(epoch) + ' ######')
    tot_correct = 0
    tot_not_na_correct = 0
    tot = 0
    tot_not_na = 0
    test_tot_correct = 0
    test_tot_not_na_correct = 0
    test_tot = 0
    test_tot_not_na = 0
    i = 0
    time_sum = 0


    while True:
        time_start = time.time()
        try:
            batch_data = train_loader.next_batch(train_loader.batch_size)
            iter_loss, iter_output, _train_op, iter_label = one_step(sess, batch_data, [loss, output, train_op_])
        except StopIteration: ## When one epoch is completed stop iteration is called
            break
        time_end = time.time()
        t = time_end - time_start
        time_sum += t
        #iter_output = iter_logit
        iter_correct = (iter_output == iter_label).sum()
        iter_not_na_correct = np.logical_and(iter_output == iter_label, iter_label != 0).sum()
        tot_correct += iter_correct
        tot_not_na_correct += iter_not_na_correct
        tot += iter_label.shape[0]
        tot_not_na += (iter_label != 0).sum()
        if tot_not_na > 0:  
            sys.stdout.write("epoch %d step %d time %.2f | loss: %f, not NA accuracy: %f, accuracy: %f\r" \
                             % (epoch, i, t, iter_loss, float(tot_not_na_correct) / tot_not_na, float(tot_correct) / tot))
            sys.stdout.flush()

        i += 1
        print("\nAverage iteration time: %f" % (time_sum / i))
        #print("Testing for early stopping ")
        try:
            test_loss,test_output,test_logit,test_label = one_step_test(sess,val_batch,[loss,output,logit])
        except StopIteration:
            pass
        #print("testing")
        test_correct = (test_output == test_label).sum()
        test_not_na_correct = np.logical_and(test_output == test_label, test_label != 0).sum()
        test_tot_correct += test_correct
        test_tot_not_na_correct += test_not_na_correct
        test_tot += test_label.shape[0]
        test_tot_not_na += (test_label != 0).sum()
        print("For testing epoch %d step %d time %.2f | loss: %f, not NA accuracy: %f, accuracy: %f\r" \
                             % (epoch, i, t, test_loss, float(test_tot_not_na_correct) / test_tot_not_na,\
                                float(test_tot_correct) / test_tot))

        #sys.stdout.flush()
        if test_loss <= best_test_loss:
            best_test_loss=test_loss
            count = 0
        else:
            count+=1
        if count > 50 and early_stopping:
            print("############################################")
            print("Early stopping at epoch %d step %d"%(epoch,i))
            print("############################################")
            break_early = True
            break

if not restore_saved_model:
    saver = tf.train.Saver()
    saver.save(sess, "./models/model_cnn2")
    print("Model Saved")
```

```python
#metrics.roc_auc_score(y_score=test_logit,y_true=pd.get_dummies(test_label,columns = np.arange(53)))
test_output = []
test_label=[]
test_logit=[]
prec = []
recall = [] 
correct = 0
test_result=[]
total_correct =0

test_loader.idx=0
counter = 0
print("############################################")
print("Testing CNN Model using testing data")
print("############################################")

while True:
    print("Running test batch  "+str(counter),end='\r')
    sys.stdout.flush()
    try:
        test_loader_batch = test_loader.next_batch(test_loader.batch_size)
        test_output,test_logit,test_label = one_step_test(sess,test_loader_batch,[output,logit])
    except StopIteration:
        break
    
    #np.append(test_logit_full,np.array(test_logit),axis=0)
    for idx in range(len(test_logit)):
        for rel in range(0, 54): 
            if test_label[idx]==rel and rel!=0: ## Skiping 0 to avoid including NAs
                test_result.append({'score': test_logit[idx][rel], 'rel_num': rel, \
                                'flag': test_output[idx]==test_label[idx], 'index': idx})
                if test_output[idx]==test_label[idx]:
                    total_correct +=1
    counter +=1
sorted_test_result = sorted(test_result, key=lambda x: x['score'])

for i, item in enumerate(sorted_test_result[::-1]):
    if item['flag']:
        correct += 1
    prec.append(float(correct) / (i + 1))
    recall.append(float(correct) / total_correct)
auc = metrics.auc(x=recall, y=prec)
```
